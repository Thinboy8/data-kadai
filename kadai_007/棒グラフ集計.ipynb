```python
!pip install japanize-matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import japanize_matplotlib

df1 = pd.read_csv("sample_pandas_6.csv")
df2 = pd.read_csv("category.csv")
df3 = pd.merge(df1, df2[['商品番号', 'カテゴリー']], how='inner', on='商品番号')

category_count = df3['カテゴリー'].value_counts()
category_count.plot(kind='bar')

plt.title('カテゴリーごとの商品数')
plt.xlabel('カテゴリー')
plt.ylabel('商品数')
plt.show()
```
